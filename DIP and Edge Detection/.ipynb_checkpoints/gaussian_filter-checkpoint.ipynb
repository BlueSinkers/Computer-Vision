This is importing all the modules in order for us to complete and use the image-detection filters.

```python
import cv2 as cv
import os
import numpy as np
from matplotlib import pyplot as plt
```

Showing the initial image (same picture in the DWT implementation in the Fourier Analysis folder)

```python
#getting the path to the image
image_path = os.path.abspath("./images/lebronslam.jpg")

#reading and processing image
image = cv.imread(image_path)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB) #converts from BGR format to RGB
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Original Image")
```

Let's run the Gaussian filter first on the output and get a more blurred version to see the impact of the Gaussian Kernel. Note we can pick the "sigma" parameter to be what we want as mentioned in my notes. Note that the parameters of the function imply we have a 21 by 21 pixel mask with the "sigma" value or standard deviation at 20. This is a bit unnecessarily blurry, but I'm just showing the effect.

```python
blurred_image = cv.GaussianBlur(image_rgb, (21,21), 20)
plt.imshow(blurred_image)
plt.axis('off')
plt.title("Gaussian Blur")
```
