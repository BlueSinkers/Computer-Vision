This is importing all the modules and images in order for us to compute the Laplacian Filter.

```python
#importing the modules
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt

#reading image and converting it from BGR to RGB
image_path = os.path.abspath("./images/lebronslam.jpg")
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#you can choose to blur the image here, but based on my experience, that just screws up the filter
```

Let's convert from the original image to the Laplacian version of the image. Note that we don't have to do much because the Laplacian is a scalar, not a vector. Also note that the result of the Laplacian outputs both negative and positive values (as one would figure), so we try to normalize and scale it. To understand more about how this system works, please check out my notes!

```python
# Apply Laplacian filter
laplacian = cv2.Laplacian(image_rgb, cv2.CV_64F)

# Convert the result to uint8
laplacian = np.uint8(np.absolute(laplacian))
```

Here I'll display the image so we can see what it looks like:

```python
#make initial image grayscale
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)

# Display the original and Laplacian-filtered images
plt.figure(figsize=(10, 5))

# Plot the original image
plt.subplot(1, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Plot the Laplacian-filtered image
plt.subplot(1, 2, 2)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian Filter')
plt.axis('off')

plt.show()
```

I made the original black-and-white so you can compare with the Laplacian edge detector. We can see how it's, to the naked eye, almost the exact same thing as the Sobel filter. BUT, they are slightly different, especially in how they're computed and how we make sense of them.
