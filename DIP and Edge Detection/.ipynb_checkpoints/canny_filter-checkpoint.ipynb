```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
#getting the path to the image
image_path = os.path.abspath("./images/lebronslam.jpg")

#reading, grayscaling, and blurring the image
image = cv2.imread(image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 1)
```

Applying Canny Edge Detector. The functions parameters are (image, threshold_1, and threshold_2) -- recall that these thresholds are based on hysteresis thresholding, and are used to classify whether something is an edge or not (read my notes to learn more). That being said, it's worth noting that the Gaussian blur kernel-size can have an impact on which values actually make sense to use here!

```python
# Apply Canny edge detector
edges = cv2.Canny(blurred_image, 100, 200)
```

```python
# Display the original image and the edge-detected image
plt.figure(figsize=(10, 5))

# Plot the original image
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Plot the edge-detected image
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edges')
plt.axis('off')

plt.show()
```
